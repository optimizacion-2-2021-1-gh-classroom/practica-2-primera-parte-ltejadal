 En este reporte probamos qué tan robusto es el algoritmo creado en la primera práctica estresándolo y probándolo con distintas configuraciones de los parámetros. Para esto probamos con distinto número de iteraciones, así como distintos números de hormigas, y nos interesa saber si podemos encontrar una distancia menor siguiendo el algoritmo de Ant Colony.
 
  Primero vamos a probar el algoritmo como quedó implementado al final de la práctica 1, vamos a hacer ejecuciones con distintas configuraciones de los parámetros, y posteriormente vamos a repetir las pruebas, pero aplicándolas a el algoritmo después de robustecerlo como parte de la segunda práctica.

```python
# importamos
#from ant_colony.aco_tsp import *
import tsplib95
import networkx as nx
from ant_colony.utils import *
from ant_colony.aco_tsp import *
from ant_colony import aco_tsp_oo
```

```python
#Cargamos el dataset
data_path2 = 'datasets/gr17.tsp'
```

```python
# lectura de datos en formato tsp
problem = tsplib95.load(data_path2)
# convertir a grafo 
G = problem.get_graph()
# matrix numerica para diccionario de distancias
G_num = nx.to_numpy_matrix(G)
# diccionario de distancias
dic_dists = create_dic_dist_from_graph(G)

# inicializacion de atraccion y feromonas
tau = init_ferom(G)
eta = init_atrac(G,dic_dists)
A = atraccion_nodos(G, tau, eta, alpha=1, beta=5)
```

## Pruebas del Algoritmo Variando Parámetros

```python
plot_graph(G, m_plot='graph')
```

```python
route, dist = hormiga_recorre(G,dic_dists, A, tau, 0, x_best=[], y_best= float('inf'))
graph_optim_path(G, route, dist)
```

```python
# antcolony
ruta, dist = ant_colony(G, dic_dists, init=0, ants=3, max_iter=500, verbose=20)
```
