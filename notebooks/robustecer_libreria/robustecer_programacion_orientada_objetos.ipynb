El propósito de este `notebook` es ejemplificar la utilización de la paquetería con su enfoque ppo aplicado a la solucíon de un problema **TSP** con la base de datos de 17 ciudades localizada en el [repositorio](https://github.com/optimizacion-2-2021-1-gh-classroom/practica-2-primera-parte-ltejadal/tree/main/datasets) y cuya fuente original está [aquí](https://people.sc.fsu.edu/~jburkardt/datasets/tsp/tsp.html).

```python
import ant_colony as ac
import time
```

```python
path = '../datasets/gr17_d_city_distances.txt'
G = ac.read_data(path)
```

Instanciar clase colonia con:
- hormigas: 2
- nodo inicial: 4
- tasa de evaporación: .5 
- alpha: 3
- beta: 3

```python
# instanciar
colonia_2 = ac.colony(G, 
                 init_node=4, 
                 n_ants=2, 
                 max_iter=10, 
                 rho=.5, 
                 alpha=3,
                 beta=3)
```

Resolver problema **TSP** y mostrar los resultados:

```python
start = time.time()
colonia_2.solve_tsp()
end = time.time()
print(f'- La mejor distancia encontrada por la colonia es: {colonia_2.best_dist}')
print(f'- La mejor ruta encontrada por la colonia es: {colonia_2.best_route}')
print(f"- Tiempo total: {(end-start)} segundos.")
```

Se mejoró la visualización de rutas sobre el grafo asociado al problema para facilitar el entendimiento de la solución. Ahora se resalta el nodo inicial, y se puede modificar el tamaño del gráfico. 

```python
colonia_2.plot_route(plt_size=(15, 8))
```
